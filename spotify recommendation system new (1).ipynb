IMPORT DEPENDENCIES

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df_reduced['combined_features'] = df_reduced['combined_features'].fillna('')
count_matrix=CountVectorizer().fit_transform(df_reduced['combined_features'])
```

```python
cs=cosine_similarity(count_matrix)
print(cs)
```

RECOMMENDER SYSTEM

```python
def return_song(track_name):
    #Arbitrary track that the user watched
    track=track_name
    #The track is
    print("The Track is:"+ track)

    track_id=df[df.Track==track]["index"].values[0]
    track_id

    scores=list(enumerate(cs[track_id]))

    sorted_scores=sorted(scores,key=lambda x:x[1],reverse=True)
    sorted_scores=sorted_scores[1:]

    i=0
    print("Top 20 similar tracks to " + track+"are\n")
    for item in sorted_scores:
        show_track=df[df.index==item[0]]["Track"].values[0]
        print(i+1,show_track)
        i=i+1
        if i>=20:
            break

    df_scores=pd.DataFrame(sorted_scores,columns=["spotify_id","score"])
```

```python
if __name__ == '__main__':
    track_name = input("Enter the TRACK: ").strip()
    song_track = return_song(track_name)

    if not song_track:
        print("No titles found.")
```
